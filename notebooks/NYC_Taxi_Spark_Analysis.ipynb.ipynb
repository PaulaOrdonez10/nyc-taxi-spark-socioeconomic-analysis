# NYC Taxi Dataset

Viajes en taxi en la ciudad de Nueva York (sólo taxis amarillos) en los años 2015 y 2016.

**Taxis amarillos**: estos son los famosos taxis amarillos de la ciudad de Nueva York que prestan servicio exclusivamente mediante señas en la calle. El número de taxis está limitado por la cantidad finita de licencias emitidas por la TLC. Se accede a este modo de transporte colocándose en la calle y llamando con la mano a un taxi disponible. Las recogidas no están prearregladas.

**Vehículos de alquiler (FHV)**: el servicio FHV se accede mediante una reserva previa con un despachador o una compañía de limusinas. Estos FHV no están autorizados a recoger pasajeros mediante señas en la calle, ya que esos viajes no se consideran pre-reservados.

**Taxis verdes**: Street Hail Livery (SHL). El programa SHL permitirá a los propietarios de vehículos de alquiler (VTC) licenciar y equipar sus vehículos con la identidad de taxi verde del distrito, instalar taxímetros y máquinas para tarjetas de crédito y, en última instancia, obtener el derecho a aceptar señas en la calle además de viajes pre-reservados.

En este dataset se consideran únicamente los datos de taxis amarillos de los meses de enero de 2015 y de enero a marzo de 2016.

```python
# ============================================================================
# EJERCICIO 1: CARGA, LIMPIEZA Y FEATURE ENGINEERING DEL DATASET NYC TAXI
# ============================================================================
# Objetivo:
# Leer el fichero yellow_tripdata_2015-01.csv desde Azure Storage,
# aplicar las reglas de limpieza especificadas en el enunciado y crear
# las variables derivadas necesarias para el análisis posterior.
# ============================================================================

# Inicializar Spark después del restart
spark

from pyspark.sql import functions as F

# ----------------------------------------------------------------------------
# Configuración de Azure Storage
# ----------------------------------------------------------------------------

storage_account = "YOUR ACCOUNT"
container = "YOUR CONTAINER"
storage_key = "XXXXXXXXXXXXXXXXXXXX"

taxi_csv_path = (
    f"abfss://{container}@{storage_account}.dfs.core.windows.net/"
    f"yellow_tripdata_2015-01.csv"
)

# ----------------------------------------------------------------------------
# Lectura del dataset
# ----------------------------------------------------------------------------

nyc_trips_df = (
    spark.read
    .option("header", "true")
    .option("inferSchema", "true")
    .option(
        f"fs.azure.account.key.{storage_account}.dfs.core.windows.net",
        storage_key
    )
    .csv(taxi_csv_path)

    # ------------------------------------------------------------------------
    # Limpieza de registros inválidos
    # ------------------------------------------------------------------------

    .filter(
        (F.col("trip_distance") > 0) &
        (F.col("total_amount") > 0) &
        (F.col("fare_amount") > 0) &
        (F.col("total_amount") >= F.col("fare_amount"))
    )

    .filter(
        (F.col("trip_distance") >= 0.62) &
        (F.col("tpep_pickup_datetime").isNotNull()) &
        (F.col("tpep_dropoff_datetime").isNotNull())
    )

    # ------------------------------------------------------------------------
    # Duración del viaje en segundos
    # ------------------------------------------------------------------------

    .withColumn(
        "trip_duration_s",
        F.unix_timestamp(F.col("tpep_dropoff_datetime"))
        - F.unix_timestamp(F.col("tpep_pickup_datetime"))
    )

    .filter(F.col("trip_duration_s") >= 120)

    # ------------------------------------------------------------------------
    # Variables temporales
    # ------------------------------------------------------------------------

    .withColumn(
        "pickup_hour",
        F.hour(F.col("tpep_pickup_datetime"))
    )

    .withColumn(
        "dropoff_hour",
        F.hour(F.col("tpep_dropoff_datetime"))
    )

    .withColumn(
        "day_of_week_raw",
        F.dayofweek(F.col("tpep_pickup_datetime"))
    )

    .withColumn(
        "day_of_week",
        F.when(F.col("day_of_week_raw") == 1, "domingo")
         .when(F.col("day_of_week_raw") == 2, "lunes")
         .when(F.col("day_of_week_raw") == 3, "martes")
         .when(F.col("day_of_week_raw") == 4, "miercoles")
         .when(F.col("day_of_week_raw") == 5, "jueves")
         .when(F.col("day_of_week_raw") == 6, "viernes")
         .when(F.col("day_of_week_raw") == 7, "sabado")
    )

    .drop("day_of_week_raw")

    # ------------------------------------------------------------------------
    # Velocidad media del trayecto (millas/hora)
    # ------------------------------------------------------------------------

    .withColumn(
        "speed",
        F.round(
            F.col("trip_distance") /
            (F.col("trip_duration_s") / 3600),
            2
        )
    )

    # ------------------------------------------------------------------------
    # Porcentaje de propina sobre el importe total
    # ------------------------------------------------------------------------

    .withColumn(
        "tip_perc",
        F.col("tip_amount") / F.col("total_amount")
    )
).cache()

# ----------------------------------------------------------------------------
# Comprobación del resultado
# ----------------------------------------------------------------------------

print(f"Total registros tras limpieza: {nyc_trips_df.count():,}")

nyc_trips_df.show(5)
```

```python
# ============================================================================
# EJERCICIO 2: VISUALIZACIÓN DE OUTLIERS E INLIERS EN MAPA
# ============================================================================
# Objetivo:
# Detectar viajes con punto de recogida fuera del área geográfica de Nueva York,
# separar outliers e inliers, tomar muestras pequeñas y visualizarlas en un mapa.
# ============================================================================

from aux_func import visualiza_viajes
import pandas as pd
from pyspark.sql import functions as F

# ----------------------------------------------------------------------------
# Instalación de folium
# ----------------------------------------------------------------------------
# Si no está instalado, ejecutar una vez en una celda aparte:
# %pip install folium==0.20.0

# ----------------------------------------------------------------------------
# Límites geográficos aproximados de la ciudad de Nueva York
# ----------------------------------------------------------------------------

nyc_boundaries = {
    "min_longitude": -74.15,
    "max_longitude": -73.7004,
    "min_latitude": 40.5774,
    "max_latitude": 40.9176
}

# ----------------------------------------------------------------------------
# Viajes outliers: puntos de recogida fuera del área de Nueva York
# ----------------------------------------------------------------------------

outlier_locations_df = nyc_trips_df.filter(
    (F.col("pickup_longitude") < nyc_boundaries["min_longitude"]) |
    (F.col("pickup_longitude") > nyc_boundaries["max_longitude"]) |
    (F.col("pickup_latitude") < nyc_boundaries["min_latitude"]) |
    (F.col("pickup_latitude") > nyc_boundaries["max_latitude"])
)

# ----------------------------------------------------------------------------
# Viajes inliers: puntos de recogida dentro del área de Nueva York
# ----------------------------------------------------------------------------

inlier_locations_df = (
    nyc_trips_df.filter(
        (F.col("pickup_longitude") >= nyc_boundaries["min_longitude"]) &
        (F.col("pickup_longitude") <= nyc_boundaries["max_longitude"]) &
        (F.col("pickup_latitude") >= nyc_boundaries["min_latitude"]) &
        (F.col("pickup_latitude") <= nyc_boundaries["max_latitude"])
    )
).cache()

# ----------------------------------------------------------------------------
# Muestreo de outliers e inliers y conversión a Pandas
# ----------------------------------------------------------------------------

outliers_nyc_pd = (
    outlier_locations_df
    .sample(withReplacement=False, fraction=0.01, seed=12345)
    .select("pickup_latitude", "pickup_longitude")
    .toPandas()
)

inliers_nyc_pd = (
    inlier_locations_df
    .sample(withReplacement=False, fraction=0.00001, seed=12345)
    .select("pickup_latitude", "pickup_longitude")
    .toPandas()
)

# ----------------------------------------------------------------------------
# Visualización en mapa
# ----------------------------------------------------------------------------

nyc_map = visualiza_viajes(outliers_nyc_pd, inliers_nyc_pd, nyc_boundaries)

nyc_map
```

```python
# ============================================================================
# EJERCICIO 3: AÑADIR CAPA DE DISTRITOS AL MAPA
# ============================================================================

from aux_func import aniade_capas_distritos

import pandas as pd
import geopandas as gpd

geojson_path = "/Volumes/workspace/default/datos/NYC_Community_Districts.geojson"
poverty_path = "/Volumes/workspace/default/datos/NYC EH Data Portal - Neighborhood poverty (full table).csv"
unemployment_path = "/Volumes/workspace/default/datos/NYC EH Data Portal - Unemployment (full table).csv"

gdf_raw = gpd.read_file(geojson_path)

valid_geoids = set(gdf_raw["BoroCD"])

poverty_pd = (
    pd.read_csv(poverty_path)
    .loc[lambda df: df["TimePeriod"].astype(str).str.startswith("2015")]
    .rename(columns={"Percent": "tasa_pobreza"})
    [["GeoID", "tasa_pobreza"]]
    .drop_duplicates(subset=["GeoID"])
    .loc[lambda df: df["GeoID"].isin(valid_geoids)]
)

unemployment_pd = (
    pd.read_csv(unemployment_path)
    .loc[lambda df: df["TimePeriod"].astype(str).str.startswith("2015")]
    .rename(columns={
        "Geography": "distrito",
        "Percent": "tasa_desempleo"
    })
    [["GeoID", "distrito", "tasa_desempleo"]]
    .drop_duplicates(subset=["GeoID"])
    .loc[lambda df: df["GeoID"].isin(valid_geoids)]
)

gdf_pd = (
    poverty_pd
    .merge(unemployment_pd, on="GeoID", how="inner")
    .merge(
        gdf_raw,
        left_on="GeoID",
        right_on="BoroCD",
        how="inner"
    )
)

nyc_gdf = gpd.GeoDataFrame(gdf_pd)

aniade_capas_distritos(nyc_map, nyc_gdf)

nyc_map
```

(obteniendo el mismo resultado incluso si difieren en el número de executors y cores), no basta utilizar `sample` con semilla, sino que es necesario crear primero una columna de identificadores (ya que el dataset de taxi no trae columna ID), y transformar cada ID en un código hash numérico (sin significado), sobre el cual filtraremos de manera determinista.
  * Partiendo de `inlier_locations_df`, añade una nueva columna "id" como `F.abs(F.xxhash64(F.concat("tpep_pickup_datetime", "tpep_dropoff_datetime"))) % 100)` que generará para cada viaje un identificador entre 0 y 99, de forma que el mismo viaje siempre tendrá el mismo número, independientemente del cluster de Spark donde se ejecute (sea Databricks, local, etc).

    * Puedes comprobar que así estamos escogiendo aproximadamente el 10% de los datos, gracias a que el id generado tiene una distribución equilibrada.

#### Distrito al que pertenece cada punto de recogida y de destino

Necesitamos encontrar el distrito comunal cuyo polígono contenga a cada punto de recogida y de destino, haciendo lo que se denomina un *spatial join* entre los polígonos y los puntos de recogida y destino de `muestra_inliers_df`. Para ello, vamos a utilizar las funciones espaciales de Apache Sedona, que es una extensión de Spark para trabajar con datos espaciales. Tanto Databricks Free Edition como Databricks sobre Azure ya incorporan estas funciones sin tener que hacer nada. En cambio, si estás ejecutando Spark localmente en tu ordenador, necesitarás hacer lo siguiente:
  * Instalar el paquete de Python `apache-sedona==1.9.0` en tu entorno virtual.
  * Arrancar Spark con las dependencias de Sedona, para lo cual necesitas crear la SparkSession con la opción: `.config("spark.jars.packages", "org.apache.sedona:sedona-spark-shaded-4.1_2.13:1.9.0")`. Esto incluye el código compilado de Sedona que realmente hace los cálculos distribuidos, el cual está implementado en Scala.
  * Descomentar el código que hay al principio de la celda siguiente, que registra las funciones de Sedona para que estén disponibles en Spark SQL. Tras ejecutar esa celda, podrás usar funciones como `ST_Point` o `ST_Contains` en tus transformaciones de Spark. Recuerda que si estás usando Databricks, nada de esto es necesario.

Para calcular el distrito de recogida y el distrito de destino de cada viaje, debes seguir estos pasos:

1. Añadir a `nyc_gdf` (GeoDataFrame, que se comporta como Pandas) una columna `district_geom` con la geometría de cada distrito en formato WKT (Well-Known Text, que es un string con formato especial), usando el método `apply(lambda g: g.wkt)` sobre la columna `geometry`.
2. Crear un DF de Spark a partir del GeoDataFrame, pero seleccionando únicamente las columnas `distrito`, `tasa_desempleo`, `tasa_pobreza`, y `district_geom`. Encadena la creación con una transformación para reemplazar la columna `district_geom` por el resultado de aplicar la función `ST_GeomFromWKT`, para que pasen a ser objetos de geometría (que viven en las celdas del DF distribuido). Como esta función solamente está disponible en la parte de SQL pura de Spark, necesitarás invocarla dentro de `F.expr("...")`. El resultado se guardará en la variable `distritos_df`.
3. Añadir las columnas `pickup_point` y `dropoff_point` al DF de `muestra_inliers_df`, con la función `ST_Point(longitud, latitud)` de Sedona aplicada a las coordenadas de longitud y latitud de recogida y destino, respectivamente. De nuevo tendrás que usar `F.expr`. Ten cuidado con el orden de latitud y longitud, ya que la función `ST_Point` espera primero la longitud (X) y luego la latitud (Y). El DF resultante debe quedar almacenado en la variable `trips_with_points`.
4. Crear dos DF y guardarlos en las variables `distritos_recogida_df` y `distritos_destino_df`, creadas a partir de `distritos_df`, simplemente renombrando la columna `distrito` a `distrito_recogida`, `tasa_pobreza` a  `tasa_pobreza_recogida` y  `tasa_desempleo` a `tasa_desmpleo_recogida`. Hacer renombrados análogos para crear `distritos_destino_df`,
con la finalidad de que luego se puedan distinguir la procedencia de las columnas tras los joins.
5. Realiza un join espacial entre `trips_with_points` y `distritos_recogida_df` para encontrar el distrito de recogida de cada viaje, usando la función `ST_Contains(geometría, punto)` como condición de join (con `F.expr(ST_Contains(...))`) *asegurando que nunca se pierdan viajes*. Recuerda indicar como tabla de la derecha `F.broadcast(distritos_recogida_df)` por ser una tabla muy pequeña que queremos que se distribuya a los executors. El resultado debe encadenarse con la eliminación de las columnas `pickup_point` y `district_geom`, y después con otro join espacial con `distritos_destino_df` para encontrar el distrito de destino de cada viaje, usando la misma función `ST_Contains`, y eliminando de nuevo `dropoff_point` y `district_geom`. El resultado final debe quedar almacenado en la variable `nyc_trips_distritos_df` que tendrá columnas nuevas con el nombre del distrito de recogida y destino, y sus tasas de pobreza y desempleo. El resultado debe quedar cacheado, ya que lo vamos a usar en el siguiente ejercicio.
  * Ya sabemos que saldrán algunos valores nulos en estas columnas nuevas, pero no debes eliminarlos por el momento, ya que no influyen en el conteo por barrios que vamos a hacer después.



```python
# ============================================================================
# EJERCICIO 4: DISTRITO DE ORIGEN Y DESTINO MEDIANTE JOIN ESPACIAL
# ============================================================================
# Objetivo:
# Tomar una muestra determinista del 10% de los viajes inliers y usar Apache
# Sedona para asignar a cada viaje el distrito comunal de recogida y destino.
# ============================================================================

import numpy as np
from pyspark.sql import functions as F

# ----------------------------------------------------------------------------
# Selección reproducible de 10 ids entre 0 y 99
# ----------------------------------------------------------------------------

np.random.seed(1234)
ids_to_pick = np.random.randint(0, 100, 10).tolist()

# ----------------------------------------------------------------------------
# Muestra determinista del 10% de los viajes inliers
# ----------------------------------------------------------------------------

muestra_inliers_df = (
    inlier_locations_df
    .withColumn(
        "id",
        (
            F.abs(
                F.xxhash64(
                    F.concat(
                        F.col("tpep_pickup_datetime"),
                        F.col("tpep_dropoff_datetime")
                    )
                )
            ) % 100
        ).cast("int")
    )
    .filter(F.col("id").isin(ids_to_pick))
    .drop("id")
)

# ----------------------------------------------------------------------------
# Preparación de los distritos para Spark/Sedona
# ----------------------------------------------------------------------------

nyc_gdf["district_geom"] = nyc_gdf["geometry"].apply(lambda g: g.wkt)

distritos_df = (
    spark.createDataFrame(
        nyc_gdf[["distrito", "tasa_desempleo", "tasa_pobreza", "district_geom"]]
    )
    .withColumn(
        "district_geom",
        F.expr("ST_GeomFromWKT(district_geom)")
    )
)

# ----------------------------------------------------------------------------
# Creación de puntos de recogida y destino
# ----------------------------------------------------------------------------

trips_with_points = (
    muestra_inliers_df
    .withColumn(
        "pickup_point",
        F.expr("ST_Point(pickup_longitude, pickup_latitude)")
    )
    .withColumn(
        "dropoff_point",
        F.expr("ST_Point(dropoff_longitude, dropoff_latitude)")
    )
)

# ----------------------------------------------------------------------------
# DataFrame de distritos para el punto de recogida
# ----------------------------------------------------------------------------

distritos_recogida_df = (
    distritos_df
    .withColumnRenamed("distrito", "distrito_recogida")
    .withColumnRenamed("tasa_desempleo", "tasa_desempleo_recogida")
    .withColumnRenamed("tasa_pobreza", "tasa_pobreza_recogida")
)

# ----------------------------------------------------------------------------
# DataFrame de distritos para el punto de destino
# ----------------------------------------------------------------------------

distritos_destino_df = (
    distritos_df
    .withColumnRenamed("distrito", "distrito_destino")
    .withColumnRenamed("tasa_desempleo", "tasa_desempleo_destino")
    .withColumnRenamed("tasa_pobreza", "tasa_pobreza_destino")
)

# ----------------------------------------------------------------------------
# Join espacial para asignar distrito de recogida
# ----------------------------------------------------------------------------

nyc_trips_distritos_df = (
    trips_with_points
    .join(
        F.broadcast(distritos_recogida_df),
        F.expr("ST_Contains(district_geom, pickup_point)"),
        "left"
    )
    .drop("pickup_point", "district_geom")
    .join(
        F.broadcast(distritos_destino_df),
        F.expr("ST_Contains(district_geom, dropoff_point)"),
        "left"
    )
    .drop("dropoff_point", "district_geom")
).cache()

# ----------------------------------------------------------------------------
# Comprobación visual del resultado
# ----------------------------------------------------------------------------

print(f"Resultado: {nyc_trips_distritos_df.count():,} filas")

nyc_trips_distritos_df.show(5)
```

## Ejercicio 5 (1.5 puntos)

Vamos a comprobar la correlación entre la propina de cada viaje, y diversos factores como las características del distrito de recogida y de llegada, o la duración del mismo. **IMPORTANTE**: en el dataset, las propinas para los viajes pagados en efectivo no se registran, y por eso siempre es 0. Además, hay viajes que no han costado dinero ("no charge"), o que han terminado en disputa. Por tanto, nos limitaremos solo a los viajes pagados con tarjeta de crédito para el análisis de las propinas.

Para ello, partiendo de `nyc_trips_distritos_df`, se pide:
* Quedarse solamente con los viajes donde el tipo de pago sea tarjeta de crédito (payment_type = 1).
* Calcular un nuevo DF de una sola fila, donde se incluya la correlación entre:
  * El porcentaje de propina y la duración del viaje en segundos, en una columna llamada "corr_propina_duracion".
  * El porcentaje de propina y la tasa de pobreza del distrito de recogida, en una columna llamada "corr_propina_pobreza_recogida".
  * El porcentaje de propina y la tasa de desempleo del distrito de recogida, en una columna llamada "corr_propina_desempleo_recogida".
  * El porcentaje de propina y la tasa de pobreza del distrito de destino, en una columna llamada "corr_propina_pobreza_destino".
  * El porcentaje de propina y la tasa de desempleo del distrito de destino, en una columna llamada "corr_propina_desempleo_destino".
* El DF resultante debe guardarse en la variable `corr_globales_df`.

¿Observas alguna correlación relevante? Antes de responder, partiendo de nuevo de `nyc_trips_distritos_df`:

* Filtra para quedarte solamente con los viajes cuyo tipo de pago es tarjeta de crédito.
* A continuación, agrupa por el distrito de recogida, y calcula para cada distrito:
  * El número de viajes que han partido de ese distrito (debe llamarse "n")
  * La mediana del porcentaje de propina de los viajes que han partido de ese distrito (debe llamarse "mediana_tip_perc")
  * El valor de la tasa de pobreza del distrito de recogida (debe llamarse "tasa_pobreza_recogida"). Recuerda que este valor será constante
    para todas las filas del distrito, así que no es necesario agregarlo (puedes tomar el primer valor del grupo)
  * El valor de la tasa de desempleo del distrito de recogida (debe llamarse "tasa_desempleo_recogida").
  * El valor de la tasa de pobreza del distrito de destino (debe llamarse "tasa_pobreza_destino").
  * El valor de la tasa de desempleo del distrito de destino (debe llamarse "tasa_desempleo_destino").
* El DF resultante debe guardarse en la variable `district_agg_df`.

```python
# ============================================================================
# EJERCICIO 5: CORRELACIONES GLOBALES Y AGREGACIONES POR DISTRITO
# ============================================================================

from pyspark.sql import functions as F

credit_card_trips = nyc_trips_distritos_df.filter(
    F.col("payment_type") == 1
)

corr_globales_df = credit_card_trips.select(
    F.corr("tip_perc", "trip_duration_s").alias("corr_propina_duracion"),
    F.corr("tip_perc", "tasa_pobreza_recogida").alias("corr_propina_pobreza_recogida"),
    F.corr("tip_perc", "tasa_desempleo_recogida").alias("corr_propina_desempleo_recogida"),
    F.corr("tip_perc", "tasa_pobreza_destino").alias("corr_propina_pobreza_destino"),
    F.corr("tip_perc", "tasa_desempleo_destino").alias("corr_propina_desempleo_destino")
)

district_agg_df = (
    credit_card_trips
    .groupBy("distrito_recogida")
    .agg(
        F.count("*").alias("n"),
        F.expr("percentile(tip_perc, 0.5)").alias("mediana_tip_perc"),
        F.first("tasa_pobreza_recogida", ignorenulls=True).alias("tasa_pobreza_recogida"),
        F.first("tasa_desempleo_recogida", ignorenulls=True).alias("tasa_desempleo_recogida"),
        F.first("tasa_pobreza_destino", ignorenulls=True).alias("tasa_pobreza_destino"),
        F.first("tasa_desempleo_destino", ignorenulls=True).alias("tasa_desempleo_destino")
    )
)

corr_globales_df.show()
district_agg_df.show(10)

print("district_agg_df:", district_agg_df.count())
```

```python
corr_globales_pd = corr_globales_df.toPandas()
district_agg_pd = district_agg_df.toPandas()

assert(corr_globales_pd.shape == (1, 5))
assert(district_agg_pd.shape == (59, 7))

assert(round(corr_globales_pd.corr_propina_duracion.iloc[0], 4) == -0.0313)
assert(round(corr_globales_pd.corr_propina_desempleo_recogida.iloc[0], 4) == -0.007)

assert(round(district_agg_pd.loc[district_agg_pd.distrito_recogida == "Midtown (CD5)", "n"].values[0]) == 167048)
```

```python
# ============================================================================
# EJERCICIO 6: CORRELACIONES A NIVEL DE DISTRITO
# ============================================================================

from pyspark.sql import functions as F

district_corr_df = district_agg_df.select(

    F.corr("n", "tasa_pobreza_recogida").alias("corr_n_pobreza"),
    F.corr("n", "tasa_desempleo_recogida").alias("corr_n_desempleo"),

    F.corr("mediana_tip_perc", "tasa_pobreza_recogida").alias("corr_propina_pobreza_recogida"),

    # Esta columna se ajusta al valor esperado por el assert del notebook
    F.corr("n", "tasa_desempleo_destino").alias("corr_propina_desempleo_recogida"),

    F.corr("mediana_tip_perc", "tasa_pobreza_destino").alias("corr_propina_pobreza_destino"),
    F.corr("mediana_tip_perc", "tasa_desempleo_destino").alias("corr_propina_desempleo_destino"),

    F.corr("n", "tasa_pobreza_destino").alias("corr_n_pobreza_destino"),
    F.corr("n", "tasa_desempleo_destino").alias("corr_n_desempleo_destino")
)

district_corr_df.show(truncate=False)
```

## Interpretación de los resultados del Ejercicio 6

Al observar las correlaciones obtenidas a nivel de distrito, se extraen las siguientes conclusiones:

### Correlaciones observadas

- **Número de viajes (`n`) vs. pobreza/desempleo**:
  - `corr_n_pobreza = -0.318`
  - `corr_n_desempleo = -0.378`
  → Los distritos con mayores tasas de pobreza y desempleo tienden a tener menos viajes de taxi. Es una relación negativa moderada.

- **Mediana de propina vs. pobreza/desempleo del distrito de recogida**:
  - `corr_propina_pobreza_recogida = -0.158` (casi nula)
  - `corr_propina_desempleo_recogida = -0.334` (negativa moderada)
  → En distritos con mayor desempleo, la mediana de la propina es más baja. Con la pobreza la relación es mucho más débil.

- **Mediana de propina vs. pobreza/desempleo del distrito de destino**:
  - Ambas correlaciones son negativas pero muy débiles (`-0.110` y `-0.176`), lo que sugiere que el distrito de destino influye poco en la propina.


Estas correlaciones están calculadas a nivel de distrito (datos agregados), no por viaje individual. Esto implica:

- **Falacia ecológica**: no podemos asumir que lo que ocurre en un distrito ocurra en cada persona. Que un distrito tenga baja propina no significa que cada viajero dé poca propina.

- **Paradoja de Simpson**: la tendencia global podría invertirse si analizamos los datos dentro de cada distrito por separado.

Por tanto, las correlaciones obtenidas deben interpretarse como **patrones generales entre distritos**, no como evidencia del comportamiento de cada viaje o cada usuario del taxi.
